## Notebook to compare Aernoet V2 / V3 inversions levels 1.5 and level 2

```python
import os
from pyaerocom import const, change_verbosity
from pyaerocom.io import ReadUngridded
OUT_DIR = './output/'
change_verbosity('warning')
if not os.path.exists(OUT_DIR):
    os.mkdir(OUT_DIR)
```

```python
read = ReadUngridded(const.AERONET_INV_V2L2_DAILY_NAME)
data_inv_v2l2 = read.read()
data_inv_v2l2.var_idx
```

```python
read = ReadUngridded(const.AERONET_INV_V3L2_DAILY_NAME)
data_inv_v3l2 = read.read()
data_inv_v3l2.contains_vars
```

```python
times_abs, abs550aer_v2, abs550aer_v3 = data_inv_v2l2.find_common_data_points(data_inv_v3l2, 'abs550aer')
times_tot, od550aer_v2, od550aer_v3 = data_inv_v2l2.find_common_data_points(data_inv_v3l2, 'od550aer')
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))
ax1.plot(abs550aer_v2, abs550aer_v3, ' *r')
```
